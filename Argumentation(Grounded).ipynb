```python
import json
import requests
# premises = ["oneEscapeRoute(f)", "lessThan4Dwellings(f)", "moreThan4Dwellings(f)"]
# premises = ["noMultiEscapeRoute(f)", "distMoreThan30(f)", "fireDoor(f)"]
premises = ["noMultiEscapeRoute(f)", "distMoreThan30(f)", "fireDoor(f)"]
rules= [
    "[r1] oneEscapeRoute(f), lessThan4Dwellings(f) => compliant(f);",
    "[r2] oneEscapeRoute(f), lobby(f), smokeVentilation(f), distLessThan10(f) => compliant(f);",
    "[r3] multiEscapeRoute(f), distLessThan30(f) => compliant(f);",
    "[r4] oneEscapeRoute(f), lobby(f), smokeVentilation(f), distLessThan10(f) => compliant(f);",
    "[r5] multiEscapeRoute(f), distLessThan30(f) => compliant(f);",
    "[r6] multiEscapeRoute(f), distLessThan30(f), fireDoor(f) => compliant(f);",
    "[r7] multiEscapeRoute(f), distLessThan30(f), distDDLessThan10(f), lobby(f), smokeVentilation(f), fireDoor(f) => compliant(f);",
    "[r8] moreThan4Dwellings(f) => noncompliant(f);",
    "[r9] noEscapeRoute(f) => noncompliant(f);",
    "[r10] distMoreThan10(f) => noncompliant(f);",
    "[r11] noMultiEscapeRoute(f) => noncompliant(f);",
    "[r12] distMoreThan30(f) => noncompliant(f);",
    "[r13] distDDMoreThan10(f) => noncompliant(f);",
    "[r14] noLobby(f) => noncompliant(f);",
    "[r15] noSmokeVentilation(f) => noncompliant(f);",
    "[r16] noFireDoor(f) => noncompliant(f);",
]
contrariness=["compliant(f)-noncompliant(f);"]
kbPrefs = []
rulePrefs = []
link = "weakest"
body = {
    "premises":premises,
    "contrariness":contrariness,
    "rules":rules,
    "link":link,
    "query":"compliant(f)",
}
headers = {"content-type":"application/json"}
res = requests.post("http://toast.arg-tech.org/api/evaluate",json=body,headers=headers)
obj = json.loads(res.text);

args = obj['arguments'];
atts = obj['defeat'];
def showOriginalArgs():
  if len(args)!= 0:
    print('Original arguments are as following');
  for arg in sorted(args): print(arg);
  if len(atts)!= 0:
     print('\nOriginal attacking relationship are as following');
  for att in sorted(atts): print(att);
  print('-'*100)


noncArgs=[];
attackers=[];

def initConclusionList():
  for arg in sorted(args):
    if 'noncompliant' in arg:
      noncArgs.append(arg);

def claimAndAsk(index):
  # print('Human: According to \"{}\", it is noncompliant.'.format(translateToUserLang(findArgumentByIndex(index))))
  # print('System: Why is \"{}\" noncompliant.'.format(translateToUserLang(findArgumentByIndex(index))))
  print('Human: According to \"{}\", it is noncompliant.'.format(findArgumentByIndex(index)))
  print('System: Why is \"{}\" noncompliant.'.format(findArgumentByIndex(index)))

def findAttacker(index):
  attackers=[];
  for att in atts:
    if index in att.split('>')[1]:
      attackers.append(att.split('>')[0]);
  # print('System: Selected \"{}\" as a attacker.'.format(translateToUserLang(findArgumentByIndex(attackers[0].split('A')[1]))))
  print('System: Selected \"{}\" as a attacker.'.format(findArgumentByIndex(attackers[0].split('A')[1])))
  return True if len(attackers)!=0 else False

def findArgumentByIndex(index):
  for arg in args:
    if 'A'+index == arg.split(':')[0]:
      return arg

def findSupportor(index):
  # print('Human: Selected \"{}\" as a supportor.'.format(translateToUserLang(findArgumentByIndex(findArgumentByIndex(index).split(':')[1].split('=')[0].split('A')[1]))));
  print('Human: Selected \"{}\" as a supportor.'.format(findArgumentByIndex(findArgumentByIndex(index).split(':')[1].split('=')[0].split('A')[1])));
  removeNoncompliant(index);


def translateToUserLang(i):
  # print('translateToUserLang\n'+i)
  if 'low(f)' in i:
    return 'Building is low.'
  if 'middle(f)' in i:
    return 'Building is middle.'
  if 'high(f)' in i:
    return 'Building is high.'
  if 'noEscapeRoute(f)' in i:
    return 'There is no escape route on this floor.'
  if 'oneEscapeRoute(f)' in i:
    return 'There is one escape route on this floor.'
  if 'multiEscapeRoute(f)' in i:
    return 'There are two or more escape route on this floor.'
  if 'noMultiEscapeRoute(f)' in i:
    return 'The number of escape route is less than 2.'
  if 'lessThan4Dwellings(f)' in i:
    return 'There are less than 4 dwellings on this floor.'
  if 'moreThan4Dwellings(f)' in i:
    return 'There are more than 4 dwellings on this floor.'
  if 'lobby(f)' in i:
    return 'There is a lobby.'
  if 'noLobby(f)' in i:
    return 'There is no lobby.'
  if 'smokeVentilation(f)' in i:
    return 'There is a smoke ventilation.'
  if 'noSmokeVentilation(f)' in i:
    return 'There is no smoke ventilation.'
  if 'distLessThan10(f)' in i:
    return 'The maximum travel distance is less than 10.'
  if 'distMoreThan10(f)' in i:
    return 'The maximum travel distance is more than 10.'
  if 'distLessThan30(f)' in i:
    return 'The maximum travel distance is less than 30.'
  if 'distMoreThan30(f)' in i:
    return 'The maximum travel distance is more than 30.'
  if 'deadEnd(f)' in i:
    return 'There is a dead end on this floor.'
  if 'noDeadEnd(f)' in i:
    return 'There is no dead end on this floor.'
  if 'fireDoor(f)' in i:
    return 'There is a fire door on this floor.'
  if 'noFireDoor(f)' in i:
    return 'There is no fire door on this floor.'
  if 'distDDLessThan10(f)' in i:
    return 'The maximum travel distance near dead end is less than 10.'
  if 'distDDMoreThan10(f)' in i:
    return 'The maximum travel distance near dead end is more than 10.'
  
  if 'noncompliant(f)' in i:
    return 'noncompliant.'
  elif 'compliant(f)' in i:
    return 'compliant.'

def removeNoncompliant(index):
  for na in noncArgs:
    if index in na:
      noncArgs.remove(na)

def showAllNoncompliant():
  print('-'*100)
  if len(noncArgs)==0: 
    print('System: I agree with human.');
  else:
    print("Here below is the noncompliant arguments list.")
    for a in noncArgs : print(a);
  print('-'*100)

def argue(index):
  claimAndAsk(index);
  findSupportor(index)
  showAllNoncompliant();  
  index=input('Please input the index of the argument in which you are interested. For example, you can input 10 to check out the reason for A10. \nInput 0 to finish. \nInput -1 to restart.')
  index = checkInput(index);
  print('-'*100)
  if index==0 or index=='0':
    print('Programme has been stopped.')
    return 
  argue(index);  

def checkInput(index):
  if index==-1 or index=='-1':
    rerun();
  if index==0 or index=='0':
    return 0;
  for x in noncArgs:
    if 'A'+index == x.split(':')[0]:
      return index
  print('Invalidated input. Please input again')
  newIndex=input('Please input the index of the argument in which you are interested. For example, you can input 10 to check out the reason for A10.')
  return checkInput(newIndex);

def showCompliantReason():
  print("Here below is the compliant arguments list.")
  for x in sorted(args):
    if 'compliant' not in x:
      print(x)

def rerun():
  noncArgs.clear();
  attackers.clear();
  main()

def main():

  # show original arguments
  # showOriginalArgs();
  print('Dialogue start:');
  # init list
  initConclusionList();
  if len(noncArgs)==0:
    print('Congratulations! There is a compliant conclusion.')
    showCompliantReason();
  else:
    showAllNoncompliant();
    index=input('Please input the index of the argument in which you are interested. For example, you can input 10 to check out the reason for A10.')
    index = checkInput(index)
    print('-'*100)
    argue(index);
    

# run main function
main();
```
